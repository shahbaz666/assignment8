```python
def guru99_exceptions():
    #importing necessary modules
    import selenium
    from selenium import webdriver as wb
    import pandas as pd
    import time

    webD = wb.Chrome('chromedriver.exe')
    webD.get('https://www.guru99.com/')
    #Click on selenium followed by selenium exceptions
    webD.find_element_by_xpath('//*[@id="java_technologies"]/li[3]/a').click()
    webD.find_element_by_xpath('//*[@id="g-mainbar"]/div/div/div/div/div/div/div[2]/table[5]/tbody/tr[34]/td[1]/a').click()
    time.sleep(5)#time for browser to load page
    #initializing lists for storing data
    names = []
    descs = []
    #finding all the row elements of the table
    table = webD.find_elements_by_tag_name('table')[3]
    rows = table.find_elements_by_tag_name('tr')
     # looping through all the row and extracting data
    for row in rows:
        name = row.find_elements_by_tag_name('td')[0].text
        desc = row.find_elements_by_tag_name('td')[1].text
        names.append(name)
        descs.append(desc)
    webD.close()
    exceptions = pd.DataFrame({'Name':names,'Description':descs})
    new_header = exceptions.iloc[0] #grab the first row for the header
    exceptions = exceptions[1:] #take the data less the header row
    exceptions.columns = new_header #set the header row as the df header
    return exceptions
```

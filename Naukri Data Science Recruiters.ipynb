```python
def naukri_ds_recruiters():
    #importing necessary modules
    import selenium
    from selenium import webdriver as wb
    import pandas as pd
    import time

    webD = wb.Chrome('chromedriver.exe')
    webD.get('https://www.naukri.com/')
    #to deal with pop up
    if len(webD.find_elements_by_xpath('//*[@id="block"]'))>0:
        webD.find_element_by_xpath('//*[@id="block"]').click()
    #click on recruiters
    webD.find_element_by_xpath('//*[@id="root"]/div[1]/div/ul[1]/li[2]/a/div').click()
    #store opened tabs on a list
    tabs = webD.window_handles
    time.sleep(4)
    #switch webdriver to find elements in the last opened tab
    webD.switch_to.window(tabs[1])
    #search for data scientist
    webD.find_elements_by_class_name('sugInp')[0].send_keys('Data Scientist')
    search = webD.find_element_by_xpath('//*[@id="qsbFormBtn"]').click()
    #initializing lists for storing data
    names = []
    companies = []
    designations = []
    locations = []
    skills = []
    #finding all the result elements of the page
    results = webD.find_elements_by_class_name('recInfo')
    # looping through all the results and extracting data
    for result in results:
        name = result.find_elements_by_tag_name('a')[1].text
        company = result.find_elements_by_tag_name('a')[2].text
        designation = result.find_elements_by_tag_name('span')[1].text
        location = result.find_elements_by_tag_name('span')[2].text
        skill = result.find_elements_by_tag_name('div')[-2].text
        names.append(name)
        companies.append(company)
        designations.append(designation)
        locations.append(location)
        skills.append(skill)
    webD.close()
    recruiters = pd.DataFrame({'Name':names,'Company':companies,'Designation':designations,'Location':locations,'Skills they hire for':skills})
    return recruiters
```

```python
print(naukri_ds_recruiters())
```
